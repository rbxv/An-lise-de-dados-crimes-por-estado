```python
#bibliotecas:
import pandas as pd #manipular dados
import matplotlib.pyplot as plt #para os gráficos
import seaborn as sns #gráficos bonitos
```

```python
#ler

df = pd.read_excel('BancoVDE 2025.xlsx') #intuitivo ler o arquivo

print("Tamanho da tabela:", df.shape) #mostra quantas linhas e colunas tem

df.head() #mostra as 5 primeiras linhas
```

```python
#analisar dados

print("Colunas", df.columns.tolist()) #motras o nome de todas as colunas

#o display é mais organizado
print("\nTipos de crime:")
display(df['evento'].unique()) #unique serve para n repetir os valores, nesse caso os tipos de crime

print("\nEstados:")
display(df['uf'].unique())

print("\nValores nulos por coluna:")
display(df.isnull().sum()) #identifica onde ta vazio

print("\nPrimeiras linhas:")
display(df.head(10)) #mostra as 5 primeiras linhas mas eu botei pra mostar 10

```

```python
#polir

df['ocorrencias'] = df['total'].fillna(df['total_vitima']) #O fillna preenche os valores nulos (NaN) com outro valor./ Lê assim: "cria a coluna ocorrencias com os valores de total, mas onde total for vazio (NaN), preenche com o valor de total_vitima"
df = df.dropna(subset=['uf', 'evento', 'ocorrencias']) #subset é bom pq n vai tirar todos os nulos só do que eu realmente quero tirar.

print("Valores nulos depois de limpar:")
display(df.isnull().sum())

print("\nTamanho depois de limpar:", df.shape)

```

```python
#polir eventos que não são crimes, vi que tavam listando eventos q n são crimes
crimes = [ #cria uma lista para poder verificar
    'Estupro',
    'Estupro de vulnerável',
    'Feminicídio',
    'Furto de veículo',
    'Homicídio doloso',
    'Lesão corporal seguida de morte',
    'Roubo a instituição financeira',
    'Roubo de carga',
    'Roubo de veículo',
    'Roubo seguido de morte (latrocínio)',
    'Tráfico de drogas',
    'Tentativa de feminicídio',
    'Tentativa de homicídio',
]

df = df[df['evento'].isin(crimes)] #verifica se o valor ta na lista


print("Eventos após filtro:")
display(df['evento'].unique()) #confirma quais eventos sobraram

print("\nTamanho da tabela após filtro:", df.shape)
```

```python
#ver quais os crimes mais cometidos

crimePorEstado = (
    df.groupby(['uf', 'evento',])['ocorrencias'] #uf e evento vao definir como agrupar e total valor que vai ser calculado
    .mean()
    .reset_index() #quando executa o groupby e o sum, o uf e evento substitui o indice(1 2 3 lista vertical), o reset_index(), "a"diciona de volta"
)

crimeMaisRecorrente = ( #tabela final
    crimePorEstado #tabela anterior com crimes por estado
    .sort_values('ocorrencias', ascending=False) # vai ordenar do maior para o menor
    .groupby('uf') #separa por estado(to muito confuso a respeito do groupby)
    .first() #pega o primeiro, como usei o sort_values vai ser o maior
    .reset_index() # reseta o indice
)

print("Crime mais recorrente por estado:")
display(crimeMaisRecorrente)

```

```python
#gráficos com matplot

plt.figure(figsize=(18, 8)) #cria area do gráfico: 18 largura e 8 de altura em polegadas

ax = sns.barplot( #chama função seaborn para criar um gráfico de barras
    data=crimeMaisRecorrente,
    x='uf', #define uf eixo X
    y='ocorrencias',# total no eixo Y
    hue='evento', #colore cada crime com uma cor diferente
    palette='tab20' #define a paleta de cores 20 cores distintas
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', fontsize=8, padding=3)

# Mais números no eixo Y
ax.yaxis.set_major_locator(plt.MaxNLocator(20))


plt.title('Crimes mais recorrentes por Estado', fontsize=16) #Titulo
plt.xlabel('Estador (UF)') #rótulo do eixo x
plt.ylabel('Total de Ocorrências') ##rótulo do eixo y
plt.xticks(rotation=0, ha='right') #Rotaciona os "uf" não vi necessidade pq n ta sobrescrevendo
plt.legend(title='Tipo de Crime', bbox_to_anchor=(1.05, 1), loc='upper left') #configura a legenda das cores, bbox: configura fora do gráfico para n cobrir as barras e loc: alinha no canto superior esquerdo
plt.tight_layout() #ajusta todo espaçamento do gráfico
plt.show() #exibe o gráfico

```
